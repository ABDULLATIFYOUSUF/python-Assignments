```python
firstValue = int(input("Enter your first number : "))
secondValue = int(input("Enter your second number : "))
operator = input("Enter operaor (+, -, x, /) : ")
if operator == "+":
    print(firstValue + secondValue)
elif operator == "-":
    print(firstValue - secondValue)
elif operator == "X":
    print(firstValue * secondValue)
elif operator == "/":
    print(firstValue / secondValue)
else:
    print("Enter valid operator")
```
